```python
import pandas as pd
```

```python
df = pd.read_csv('./station_location.txt', header=0, delim_whitespace=True)
```

```python
strings = ''
for ix, row in df.iterrows():
    
    strings += '{lat: %.6f, lng: %.6f},\n'%(row.latitude, row.longitude)
```

```python
text_file = open("GPS_locations.txt", "w")
text_file.write(strings)
text_file.close()
```
